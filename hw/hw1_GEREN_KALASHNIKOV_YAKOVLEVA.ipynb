# Support functions

```python
def preprocess_string(s: str) -> str:
    return strip_short(strip_punctuation(remove_stopwords(any2unicode(s)))).split()

def tokenize_string(s: str) -> list:
    return [*tokenize(s)]
```

```python
preprocess_string("I'm in your room")
```

```python
tokenize_string("I'm in your room")
```

```python
class fileloader:
    def __init__(self, pathdir, processing=preprocess_string):
        self.pathdir = pathdir
        self.processing = processing
    
    def __iter__(self):
        filenames = [filename for filename in os.listdir(self.pathdir)]

        for fname in filenames:
            with open(self.pathdir + fname, 'r') as f:
                text = f.read()
            yield self.processing(text)
```

## Обработка данных

```python
train_df = pd.read_csv('./hpac_corpus/hpac_training_128.tsv', sep = '\t', header = None)
test_df = pd.read_csv('./hpac_corpus/hpac_test_128.tsv', sep = '\t', header = None)
val_df = pd.read_csv('./hpac_corpus/hpac_dev_128.tsv', sep = '\t', header = None)
```

```python
def make_dataset(df, id2label, max_len=128, emb_dim=100, pad_value=0, preprocess=preprocess_string):
    dataset = []
    
    for i, row in tqdm(df.iterrows(), total=len(df)):
        label, text = row[1], row[2]
        preprocessed = preprocess(text)
        data = np.full((max_len, emb_dim), pad_value, dtype=np.float32)
        
        for i, word in enumerate(preprocessed):
            if i == max_len:
                break
                
            data[i] = lm_model.wv[word]

        label_id = id2label.index(label)
        dataset.append([data, label_id])
    return dataset
```

```python
id2label = sorted(list(train_df[1].unique()))

train_dataset = make_dataset(train_df, id2label)
val_dataset = make_dataset(val_df, id2label)
test_dataset = make_dataset(test_df, id2label)
```

## Обучение бейзлайна
* Используется линейная модель с дропаутом

```python
X_train = np.array([item[0].mean(axis=0) for item in train_dataset], dtype=np.float32)
X_test = np.array([item[0].mean(axis=0) for item in test_dataset], dtype=np.float32)
X_val = np.array([item[0].mean(axis=0) for item in val_dataset], dtype=np.float32)
```

```python
Y_train = np.array([item[1] for item in train_dataset])
Y_test = np.array([item[1] for item in test_dataset])
Y_val = np.array([item[1] for item in val_dataset])
```

```python
class BaselineNet(nn.Module):
    def __init__(self, num_units=1024, nonlin=nn.ReLU()):
        super().__init__()

        self.dense0 = nn.Linear(100, num_units)
        self.nonlin = nonlin
        self.dropout = nn.Dropout(0.4)
        self.dense1 = nn.Linear(num_units, num_units)
        self.output = nn.Linear(num_units, 85)
        self.softmax = nn.Softmax(dim=-1)

    def forward(self, X, **kwargs):
        X = self.nonlin(self.dense0(X))
        X = self.dropout(X)
        X = self.nonlin(self.dense1(X))
        X = self.softmax(self.output(X))
        return X
```

```python
valid_ds = Dataset(X_val, Y_val)
f1 = EpochScoring(scoring='f1_macro', lower_is_better=False)

net = NeuralNetClassifier(
    BaselineNet,
    max_epochs=100,
    lr=0.1,
    iterator_train__shuffle=True,
    device="cuda",
    train_split=predefined_split(valid_ds),
    callbacks=[f1],
    batch_size=64
)

net.fit(X_train, Y_train)
```

```python
Y_pred = net.predict(X_test)
f1_score(Y_test, Y_pred, average='macro')
```

## Обучение модели CNN + k-max pooling.

```python
def get_out_size(sequence_length, filter_size, padding=0, dilation=1, stride=1):
    length = sequence_length + 2 * padding - dilation * (filter_size - 1) - 1
    length = int(length / stride)
    return length + 1

def global_max_pooling(tensor, k):
    ret, _ = torch.topk(tensor, k)
    return ret


class CNN(nn.Module):
    def __init__(self, augs=False):
        super().__init__()
        
        params = {
            "k": None,
            'out_channels' : 512,
            'kernel_sizes' : [4, 8, 16],
            'hidden_dim1' : 1024,
            'conv_stride' : 2,
            'pool_stride' : 8,
            "embedding_dim": 100,
            "sequence_length": 128,
            "classes": 85,
        }
        
        self.k = params["k"]
        self.kernel_sizes = params['kernel_sizes']
        self.kernels_num = len(params['kernel_sizes'])
        self.out_size = 0

        self.conv_layers = nn.ModuleList()
        self.pool_layers = nn.ModuleList()
        
        self.augs = augs
    

        for fsz in self.kernel_sizes :        
            conv = nn.Conv1d(params['embedding_dim'], params['out_channels'], fsz, stride=params['conv_stride'])
            pool = nn.MaxPool1d(params['pool_stride'], stride=params['pool_stride'])
            out_size = get_out_size(params['sequence_length'], fsz, stride=params['conv_stride'])
            pool_out_size = int(out_size * params['out_channels'] / params['pool_stride']) 
            self.out_size += pool_out_size

            self.conv_layers.append(conv)
            self.pool_layers.append(pool)
        
        if self.k is not None:
            self.linear1 = nn.Linear(self.kernels_num * self.k, params['hidden_dim1'])
        else:
            self.linear1 = nn.Linear(self.out_size, params['hidden_dim1'])
            
        self.linear2 = nn.Linear(params['hidden_dim1'], params['classes'])
        
        # init 
        for conv in self.conv_layers:
            torch.nn.init.xavier_uniform_(conv.weight)
        torch.nn.init.xavier_uniform_(self.linear1.weight)
        torch.nn.init.xavier_uniform_(self.linear2.weight)


    def forward(self, texts):
        if self.augs:
            rand_word = np.random.choice(lm_model.wv.index_to_key)
            texts[:, torch.randperm(80)[:15], :] = torch.FloatTensor(lm_model.wv[rand_word]).cuda()
        
        texts = texts.permute(0, 2, 1)
        
        conv_out = []

        for i in range(len(self.kernel_sizes)) :
            outputs = self.conv_layers[i](texts)
            outputs = outputs.view(outputs.shape[0], 1, outputs.shape[1] * outputs.shape[2])
            outputs = F.relu(outputs)
            if self.k is not None:
                outputs = global_max_pooling(outputs, self.k)
            else:
                outputs = self.pool_layers[i](outputs)
            outputs = outputs.view(outputs.shape[0], -1)
            conv_out.append(outputs)
            del outputs

        if len(self.kernel_sizes) > 1 :
            outputs = torch.cat(conv_out, 1)
        else:
            outputs = conv_out[0]
        
        outputs = self.linear1(outputs)
        outputs = F.relu(outputs)
        outputs = nn.Dropout(p=0.4)(outputs)
        outputs = self.linear2(outputs)
        outputs = nn.Softmax(dim=-1)(outputs)
        return outputs
```

```python
X_train = np.array([item[0] for item in train_dataset], dtype=np.float32)
X_val = np.array([item[0] for item in val_dataset], dtype=np.float32)
X_test = np.array([item[0] for item in test_dataset], dtype=np.float32)
```

```python
valid_ds = Dataset(X_val, Y_val)
f1 = EpochScoring(scoring='f1_macro', lower_is_better=False)
```

```python
net = NeuralNetClassifier(
    module=CNN,
    module__augs=False,
    max_epochs=300,
    lr=0.01,
    device="cuda:0",
    train_split=predefined_split(valid_ds),
    callbacks=[f1],
    batch_size=128,
)

net.fit(X_train, Y_train)
```

```python
Y_pred = net.predict(X_test)
f1_score(Y_test, Y_pred, average='macro')
```

Аугментация с помощью случайной замены слов в данных не дала прироста в итоговом скоре, что может быть связано как с общими недочетами модели, так и с тем, что в данной задаче была применена обширная предобработка данных, что заведомо снижает эффективность аугментации.
